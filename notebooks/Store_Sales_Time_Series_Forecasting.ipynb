```python
import pandas as pd

train_df = pd.read_csv("StoreSalesData/train.csv")
test_df = pd.read_csv("StoreSalesData/test.csv")
sample_submission = pd.read_csv("StoreSalesData/sample_submission.csv")

print("‚úÖ Data loaded successfully!")
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# Preview the training data
train_df.head()
```

```python
# Feature Engineering

# Convert date again just in case
train_df['date'] = pd.to_datetime(train_df['date'])
test_df['date'] = pd.to_datetime(test_df['date'])

# Date-based features
for df in [train_df, test_df]:
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['dayofweek'] = df['date'].dt.dayofweek
    df['weekofyear'] = df['date'].dt.isocalendar().week

# Label encode store_nbr and family
from sklearn.preprocessing import LabelEncoder

le_family = LabelEncoder()
le_store = LabelEncoder()

train_df['family'] = le_family.fit_transform(train_df['family'])
test_df['family'] = le_family.transform(test_df['family'])

train_df['store_nbr'] = le_store.fit_transform(train_df['store_nbr'])
test_df['store_nbr'] = le_store.transform(test_df['store_nbr'])

print("üéâ Feature Engineering Completed!")
```

```python
# Step 3: Create Lag & Rolling Features
# Sort before creating lag features
train_df = train_df.sort_values(by=["store_nbr", "family", "date"])

# Group by store and family
group = train_df.groupby(['store_nbr', 'family'])

# Lag features
train_df['lag_7'] = group['sales'].shift(7)
train_df['lag_14'] = group['sales'].shift(14)

# Rolling mean features
train_df['rolling_mean_7'] = group['sales'].shift(1).rolling(7).mean()
train_df['rolling_mean_14'] = group['sales'].shift(1).rolling(14).mean()

# Fill missing values with 0
train_df.fillna(0, inplace=True)

print("‚úÖ Lag and Rolling Features Created")
```

```python
# Step 4: Model Prep and Train XGBoost
# Features and target
features = ['store_nbr', 'family', 'onpromotion', 'year', 'month', 'day', 'dayofweek', 'weekofyear',
            'lag_7', 'lag_14', 'rolling_mean_7', 'rolling_mean_14']
target = 'sales'

# Filter only training data with lag features
train_filtered = train_df[train_df['date'] < '2017-08-01']  # Just before test starts
valid_filtered = train_df[train_df['date'] >= '2017-08-01']

# X and y
X_train = train_filtered[features]
y_train = train_filtered[target]
X_valid = valid_filtered[features]
y_valid = valid_filtered[target]

# XGBoost
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=100, max_depth=8, learning_rate=0.1, subsample=0.8, random_state=42)
model.fit(X_train, y_train)

print("‚úÖ Model trained!")
```

```python
# Step 5: Evaluate the Model
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = model.predict(X_valid)
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))

print(f"üìâ Validation RMSE: {rmse:.2f}")
```

```python
import pandas as pd
# Step 6: Predict on Test Data
# Merge last few days of train with test to create lag features
full_df = pd.concat([train_df, test_df], sort=False)
full_df = full_df.sort_values(by=['store_nbr', 'family', 'date'])

# Recompute lag features
group = full_df.groupby(['store_nbr', 'family'])
full_df['lag_7'] = group['sales'].shift(7)
full_df['lag_14'] = group['sales'].shift(14)
full_df['rolling_mean_7'] = group['sales'].shift(1).rolling(7).mean()
full_df['rolling_mean_14'] = group['sales'].shift(1).rolling(14).mean()
full_df.fillna(0, inplace=True)

# Get test set with features
test_final = full_df[full_df['date'] >= '2017-08-15']
X_test = test_final[features]

# Predict
test_final['sales'] = model.predict(X_test)

# Prepare for submission
submission = test_final[['id', 'sales']]
submission.to_csv("xgb_submission.csv", index=False)
print("‚úÖ Submission file created!")
```

```python
# Create a clean DataFrame with necessary columns
final_output = test_final[['id', 'date', 'store_nbr', 'family']].copy()
final_output['predicted_sales'] = test_final['sales']

# Save as CSV to upload in Power BI
final_output.to_csv("PowerBI_Sales_Predictions.csv", index=False)

print("‚úÖ Power BI input CSV file created successfully.")
```
